```python
import matplotlib.pyplot as plt
import pylab
import numpy as np
import torch
import os
os.getcwd()
```

```python
img = plt.imread("yy.jpg")[:, :, 0]
plt.imshow(img, cmap='gray')
pylab.show()
```

```python
print(img.shape)
img = torch.from_numpy(img)
```

```python
fil = torch.tensor([[-1, -1, 0],
                   [-1, 0,  1],
                   [ 0, 1, 1]])
```

```python
fil_heigh = fil.shape[0]
fil_width = fil.shape[1]
conv_heigh = img.shape[0] - fil.shape[0] + 1
conv_width = img.shape[1] - fil.shape[1] + 1
```

```python
conv = torch.zeros((conv_heigh, conv_width),dtype = torch.uint8)
for i in range(conv_heigh):
    for j in range(conv_width):
        conv[i][j] = torch.mul(img[i:i + fil_heigh, j:j +fil_width],fil).sum()
```

```python
conv = conv.numpy()
plt.imshow(conv, cmap='gray')
pylab.show()
```
