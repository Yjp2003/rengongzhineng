## 使用深度学习实现Kaggle房价预测

```python
import pandas as pd
import torch
```

### 1. 加载Kaggle房价数据

```python
train_d = pd.read_csv("./train.csv")
test_d = pd.read_csv("./test.csv")
```

```python
train_d
```

```python
test_d
```

```python
X_label = train_d.iloc[:, -1]
X_label
```

```python
X_train = train_d.iloc[:, 1:-1]
X_test = test_d.iloc[:, 1:]
print("train shape for X:", X_train.shape)
print("test shape for X:", X_test.shape)
all_features = pd.concat([X_train, X_test])
all_features
```

```python
all_features.describe()
```

```python
all_features.info()
```

```python
#统计缺失值
miss_value = pd.DataFrame(data=all_features.isna().sum().values, index=all_features.isna().sum().index, columns=["缺失值"])
miss_value = miss_value[miss_value["缺失值"]>0].sort_values("缺失值", ascending=False)
miss_value["缺失率"] = miss_value["缺失值"]/all_features.shape[0]
miss_value
```

### 2. 数据预处理

```python
all_features.drop(["PoolQC", "MiscFeature", "Alley"], axis=1, inplace=True)
```

```python
all_features
```

```python
#用0填充缺失值
all_features.fillna(0, inplace=True)
all_features
```

```python
#对数值型特征进行z-score处理
numerical_col = all_features.dtypes[all_features.dtypes!="object"].index
all_features[numerical_col] = all_features[numerical_col].apply(lambda x: (x-x.mean())/x.std())
all_features
```

```python
#对类型的列进行独热编码处理
new_features = pd.get_dummies(all_features, dtype=int)
new_features
```

```python
#特征数据转换成张量数据
nn = X_train.shape[0]
new_train_X = torch.tensor(new_features[:nn].values, dtype=torch.float32)
new_test_X = torch.tensor(new_features[nn:].values, dtype=torch.float32)
train_labels = torch.tensor(X_label.values.reshape(-1, 1), dtype=torch.float32) # 取出价格标签
```

### 3.构建深度学习模型

```python
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
#超参数
input_size = new_train_X.shape[1]
output_size = 1
learning_rate = 0.001
num_epochs = 1000
```

```python
#线性回归模型
model = nn.Linear(input_size, output_size)
#损失函数和优化器
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
#训练模型
loss_d = []
for num in range(num_epochs):
    inputs = new_train_X
    targets = train_labels
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    #反向传播
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    loss_d.append(loss.item())

    if (num+1)%5 == 0:
        print("Epoch %d/%d, Loss: %.4f"%(num+1, num_epochs, loss.item()))
```

```python
#画图
plt.plot(range(num_epochs), loss_d)
plt.xlabel("Epoch")
plt.ylabel("Loss Value")
plt.show()
```

```python
#对测试集进行预测
model(new_test_X)
```

```python
# 保存模型参数
torch.save(model.state_dict(), 'model.ckpt')
```
